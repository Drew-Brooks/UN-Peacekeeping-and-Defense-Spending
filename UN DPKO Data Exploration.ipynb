## Adding Defense Spending Data

```python
# Add defense spending data
# Spending in %GDP, downloaded from https://data.worldbank.org/indicator/MS.MIL.XPND.GD.ZS 
# I had to delete the first three rows before loading it into the Jupyter folder because they were just summarizing the source and as-of date
spend = pd.read_csv('WB_def_exp.csv', delimiter=',') 
spend.head(5)
```

```python
spend.info()
```

```python
pcw = spend.iloc[:, 34:63]
pcw.describe()
```

```python
# rename col "Country Code" to "Contributor_ISO-3"
spend.rename(columns={'Country Code':'Contributor_ISO-3'}, inplace=True)
```

```python
# select relevant columns (ISO-3 and 1990-2019) and melt
spend=spend.filter(['Contributor_ISO-3','1990','1991','1992','1993','1994','1995','1996','1997','1998','1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020'])
spend=spend.melt('Contributor_ISO-3', var_name='Year', value_name='DefSpendGDP%')
spend['Year']=spend['Year'].astype(int)
spend.head(5)
```

```python
plt.figure(figsize=(16, 16))
ax = sns.boxplot(x="Year", y="DefSpendGDP%", data=spend)
```

```python
print(spend.loc[spend['DefSpendGDP%'] > 20])
```

```python
# None of the countries with spending above 20% of GDP are PKO contributors, so I can lose that
spend = spend[spend['DefSpendGDP%'] < 20]
```

```python
plt.figure(figsize=(16, 10))
ax = sns.boxplot(x="Year", y="DefSpendGDP%", data=spend)
```

```python
# add year column to df
df['Year'] = pd.DatetimeIndex(df['Date']).year
df.head(5)
```

```python
# merge spending by year
dfs=df.merge(spend, left_on=['Contributor_ISO-3','Year'], right_on=['Contributor_ISO-3','Year'], how='left')
dfs.head(5)
```

```python
plt.figure(figsize=(16, 10))
sns.scatterplot(data=dfs, x="DefSpendGDP%", y="Total")
```

```python
plt.figure(figsize=(16, 10))
sns.scatterplot(data=dfs, x="DefSpendGDP%", y="Total", hue="Contributor_Continent")
```

```python
plt.figure(figsize=(16, 10))
sns.scatterplot(data=dfs, hue="DefSpendGDP%", y="Total", x="Year")
```

```python
plt.figure(figsize=(16, 10))
sns.scatterplot(data=dfs, hue="Contributor_Continent", y="Total", x="Year")
```

```python
## Save off df as csv so I can use it in other notebooks
dfs.to_csv(r'contribsandspend.csv', index='false')
```
